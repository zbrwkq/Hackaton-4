```python
# Assuming 'df' is your DataFrame
medal_counts_reset['game_year'] = medal_counts_reset['game_year'].astype(int)
# medal_counts_reset['game_season'] = medal_counts_reset['game_season'].astype(str)
medal_counts_reset['game_location'] = medal_counts_reset['game_location'].astype(str)
medal_counts_reset['country_name'] = medal_counts_reset['country_name'].astype(str)
medal_counts_reset.dtypes
```

```python
# Example using manual steps for preprocessing and model training
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Data preparation
X = medal_counts_reset[['game_year', 'country_name']]
y = medal_counts_reset[['BRONZE', 'GOLD', 'SILVER']]
encoder = OneHotEncoder(handle_unknown='ignore')
X_encoded = encoder.fit_transform(X)

# Model training
model = RandomForestRegressor(n_estimators=100, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
model.fit(X_train, y_train)

# from sklearn.preprocessing import OneHotEncoder
# from sklearn.ensemble import RandomForestRegressor
# from sklearn.model_selection import train_test_split
# import pandas as pd
# import numpy as np

# # Load your data into a DataFrame
# # medal_counts_reset should already be loaded and contain the relevant columns

# # Prepare the feature and target arrays
# X = medal_counts_reset[['game_year', 'game_location', 'country_name']]
# y = medal_counts_reset[['BRONZE', 'GOLD', 'SILVER']]

# # Encoding categorical data
# encoder = OneHotEncoder(handle_unknown='ignore')
# X_encoded = encoder.fit_transform(X)  # This returns a sparse matrix

# # Converting sparse matrix to a dense NumPy array format
# X_dense = np.asarray(X_encoded.todense())

# # Splitting the dataset into training and testing sets
# X_train, X_test, y_train, y_test = train_test_split(X_dense, y, test_size=0.2, random_state=42)

# # Creating and training the RandomForestRegressor
# model = RandomForestRegressor(n_estimators=100, random_state=42)
# model.fit(X_train, y_train)

```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
# Make predictions on the test data
y_pred = model.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Calculate mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

# Calculate R-squared value
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'medal_counts_reset2' is your DataFrame
# Group by 'game_year' and 'discipline_title', and count occurrences
discipline_counts = medal_counts_reset2.groupby(['game_year', 'discipline_title']).size()

# Convert the series to a DataFrame
discipline_counts = discipline_counts.reset_index(name='counts')

# Sort the data by year and count to see the least frequent disciplines at the top for each year
discipline_counts_sorted = discipline_counts.sort_values(by=['game_year', 'counts'], ascending=[True, True])

# Optionally, filter to see the least common disciplines per year
# Here you might want to show, for example, the least 5 disciplines per year
filtered_disciplines = discipline_counts_sorted.groupby('game_year').head(5)

# Plotting
plt.figure(figsize=(12, 8))
for year, group in filtered_disciplines.groupby('game_year'):
    plt.barh(group['discipline_title'], group['counts'], label=f'Year {year}')
plt.xlabel('Count of Disciplines')
plt.ylabel('Disciplines')
plt.title('Least Common Disciplines per Year')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'medal_counts_reset2' is your DataFrame
# Step 1: Calculate the total medals awarded each year
total_medals_per_year = medal_counts_reset2.groupby('game_year').size()

# Step 2: Calculate counts for each discipline per year
discipline_counts = medal_counts_reset2.groupby(['game_year', 'discipline_title']).size().reset_index(name='counts')

# Step 3: Calculate the probability of each discipline appearing each year
discipline_counts['probability'] = discipline_counts.apply(lambda row: row['counts'] / total_medals_per_year[row['game_year']], axis=1)

# Step 4: Plot these probabilities for each discipline across years
unique_disciplines = discipline_counts['discipline_title'].unique()

plt.figure(figsize=(15, 10))
for discipline in unique_disciplines:
    data = discipline_counts[discipline_counts['discipline_title'] == discipline]
    plt.plot(data['game_year'], data['probability'], marker='o', label=discipline)

plt.xlabel('Year')
plt.ylabel('Probability')
plt.title('Probability of Discipline Occurrence per Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
# from sklearn.ensemble import GradientBoostingRegressor


import joblib
medal_counts_reset2.drop_duplicates(subset=['game_year', 'game_location', 'country_name', 'discipline_title', 'event_title'], inplace=True)
# Preparing the data
medal_counts_reset2 = medal_counts_reset2.loc[medal_counts_reset2['country_name'] != 'None']
X = medal_counts_reset2[['game_year', 'game_location', 'country_name', 'discipline_title', 'event_title']]
y = medal_counts_reset2[['BRONZE', 'GOLD', 'SILVER']]

encoder = OneHotEncoder(handle_unknown='ignore')
X_encoded = encoder.fit_transform(X)
# Save the encoder
joblib.dump(encoder, 'encoder2.h5')
# Model training
model = RandomForestRegressor(n_estimators=100, random_state=42)
# model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Assuming X_test and y_test are already defined and the model is trained
y_pred = model.predict(X_test)

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
```
