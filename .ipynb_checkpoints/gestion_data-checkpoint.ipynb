```python
def clean_and_convert(df):
    # Loop over each column in the DataFrame
    for column in df.columns:
        # If the column contains only numeric values or NaNs, convert to float then to int if possible
        if pd.to_numeric(df[column], errors='coerce').notna().all():
            df[column] = pd.to_numeric(df[column], errors='coerce').fillna(0)
            # Convert float to int if no decimal part is present
            if (df[column] % 1 == 0).all():
                df[column] = df[column].astype(int)
        else:
            # For columns with non-numeric strings
            try:
                # First, try converting to datetime
                df[column] = pd.to_datetime(df[column])
                continue  # Skip further processing if datetime conversion is successful
            except (ValueError, TypeError):
                # If not datetime, treat as string
                df[column] = df[column].fillna('NULL').astype(str)
```
