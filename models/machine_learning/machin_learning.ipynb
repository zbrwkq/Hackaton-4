```python
merged_df = merged_df.loc[merged_df['medal_type'] != 'None']
```

```python
merged_q1_df = merged_df[['game_year', 'game_location', 'country_name', 'medal_type']]
merged_q1_df = merged_q1_df.loc[merged_q1_df['country_name'] != 'None']
```

```python
# Filter the DataFrame to include only the medal types of interest
# filtered_df = merged_q1_df[merged_q1_df['medal_type'].isin(['GOLD', 'SILVER', 'BRONZE'])]
# filtered_df = merged_q1_df['medal_type'].isin(['GOLD', 'SILVER', 'BRONZE'])
# Perform the grouping and count the number of each medal type for each combination of game_year and game_location
medal_counts = merged_q1_df.groupby(['game_year', 'game_location', 'country_name', 'medal_type']).size().unstack(fill_value=0)

# This will result in a DataFrame where each row represents a (game_year, game_location) combination,
# and columns represent counts for 'GOLD', 'SILVER', and 'BRONZE'.
medal_counts
```

```python
medal_counts_reset = medal_counts.reset_index()
# medal_counts_reset = medal_counts_reset.drop(medal_counts_reset.columns[0], axis=1)
medal_counts_reset
```

```python
# Assuming 'df' is your DataFrame
medal_counts_reset['game_year'] = medal_counts_reset['game_year'].astype(int)
medal_counts_reset['game_location'] = medal_counts_reset['game_location'].astype(str)
medal_counts_reset['country_name'] = medal_counts_reset['country_name'].astype(str)
medal_counts_reset.dtypes
```

```python
# Example using manual steps for preprocessing and model training
%pip install scikit-learn
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Data preparation
X = medal_counts_reset[['game_year', 'game_location', 'country_name']]
y = medal_counts_reset[['BRONZE', 'GOLD', 'SILVER']]
encoder = OneHotEncoder(handle_unknown='ignore')
X_encoded = encoder.fit_transform(X)

# Model training
model = RandomForestRegressor(n_estimators=100, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
# Make predictions on the test data
y_pred = model.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Calculate mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

# Calculate R-squared value
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)

```

```python
import joblib

# Save the model
joblib.dump(model, 'medal_predictor_model.h5')
joblib.dump(encoder, 'encoder.h5')
```

```python
import joblib
import numpy as np

# Load the model
loaded_model = joblib.load('medal_predictor_model.h5')

# Preparing new data for prediction
new_data = pd.DataFrame({
    'game_year': [2022],
    'game_location': ['France'],  # Suppose 'France' was encoded as '8' in your original model
    'country_name': ['France']    # Suppose 'France' was encoded as '43' in your original model
})

# Encoding the new data using the same encoder used during training
new_data_encoded = encoder.transform(new_data)

# Predicting with the model
predictions = loaded_model.predict(new_data_encoded)
rounded_predictions = np.round(predictions).astype(int)
print("Predicted medal counts (Bronze, Gold, Silver):", rounded_predictions)
```
