```python
medal_counts = df['country_name'].value_counts()
medal_counts = medal_counts.head(20)

plt.figure(figsize=(10, 6))
plt.bar(medal_counts.index, medal_counts.values, color='skyblue')
plt.xlabel('Pays')
plt.ylabel('Nombre de médailles')
plt.title('Distribution des médailles par pays')
plt.xticks(rotation=90)
plt.show()
```

```python
total_medals_by_country = df['country_name'].value_counts()

countries_with_over_100_medals = total_medals_by_country[total_medals_by_country > 100].index

filtered_df = df[df['country_name'].isin(countries_with_over_100_medals)]

filtered_df['year'] = filtered_df['slug_game'].str.extract(r'(\d{4})').astype(int)

medals_by_country_and_year = filtered_df.groupby(['year', 'country_name']).size().unstack(fill_value=0)

medals_by_country_and_year = medals_by_country_and_year.sort_index()

plt.figure(figsize=(14, 8))
for country in medals_by_country_and_year.columns:
    plt.plot(medals_by_country_and_year.index, medals_by_country_and_year[country], marker='o', label=country)

plt.xlabel('Année')
plt.ylabel('Nombres de médailles')
plt.title('Evolution des Médailles pour chaque Edition des Jeux Olympiques')
plt.legend(title='Pays')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### On remarque les irrégularités liés au JO été et hiver

```python
merged_df = pd.merge(df, df_hosts, left_on='slug_game', right_on='game_slug', how='inner')
merged_df.set_index('id', inplace=True)
merged_df.head()
```

```python
merged_df.drop(['country_code', 'country_3_letter_code', 'game_slug', 'game_start_date', 'game_start_date'], axis=1, inplace=True)
df_winter = merged_df[merged_df['game_season'] == 'Winter']
df_summer = merged_df[merged_df['game_season'] == 'Summer']
```

```python
medal_counts = df_summer['country_name'].value_counts()
medal_counts = medal_counts.head(20)

plt.figure(figsize=(10, 6))
plt.bar(medal_counts.index, medal_counts.values, color='skyblue')
plt.xlabel('Pays')
plt.ylabel('Nombre de médailles')
plt.title('Distribution des médailles par pays pour les JO d\'été')
plt.xticks(rotation=90)
plt.show()
```

```python
medal_counts = df_winter['country_name'].value_counts()
medal_counts = medal_counts.head(20)

plt.figure(figsize=(10, 6))
plt.bar(medal_counts.index, medal_counts.values, color='skyblue')
plt.xlabel('Pays')
plt.ylabel('Nombre de médailles')
plt.title('Distribution des médailles par pays pour les JO d\'hiver')
plt.xticks(rotation=90)
plt.show()
```

```python
count = merged_df['game_season'].value_counts()
count = count.head(20)

plt.figure(figsize=(10, 6))
plt.bar(count.index, count.values, color='skyblue')
plt.xlabel('Saison')
plt.ylabel('Nombre de médailles possibles')
plt.title('Nombre d\'épreuve en fonction de la saison')
plt.xticks(rotation=90)
plt.show()
```

### La différence de nombre de médaille entre les JO d'hiver et d'été s'explique par la différence du nombre d'épreuve

```python
medals = df_summer.groupby(['country_name', 'medal_type']).size().unstack(fill_value=0)
medals['Total'] = medals.sum(axis=1)
medals = medals.sort_values(by='Total', ascending=False).head(10)

ax = medals[['GOLD', 'SILVER', 'BRONZE']].plot(kind='bar', stacked=False, figsize=(12, 8), color=['#FFD700', '#C0C0C0', '#CD7F32'])

plt.xlabel('Pays')
plt.ylabel('Nombre de médailles')
plt.title('Nombre de médailles des 10 pays ayant le plus de médailles par type de médaille (été)')
plt.legend(title='Type de médaille', loc='upper right')
plt.grid(True)
plt.show()
```

```python
medals = df_winter.groupby(['country_name', 'medal_type']).size().unstack(fill_value=0)
medals['Total'] = medals.sum(axis=1)
medals = medals.sort_values(by='Total', ascending=False).head(10)

ax = medals[['GOLD', 'SILVER', 'BRONZE']].plot(kind='bar', stacked=False, figsize=(12, 8), color=['#FFD700', '#C0C0C0', '#CD7F32'])

plt.xlabel('Pays')
plt.ylabel('Nombre de médailles')
plt.title('Nombre de médailles des 10 pays ayant le plus de médailles par type de médaille (hiver)')
plt.legend(title='Type de médaille', loc='upper right')
plt.grid(True)
plt.show()
```

```python
medals_by_country_and_year = df_summer.groupby(['country_name', 'game_year']).size().unstack(fill_value=0)

top_countries = df['country_name'].value_counts().head(10).index

medals_by_country_and_year = medals_by_country_and_year.loc[top_countries]

plt.figure(figsize=(12, 8))
for country in medals_by_country_and_year.index:
    plt.plot(medals_by_country_and_year.columns, medals_by_country_and_year.loc[country], label=country)

plt.xlabel('Année')
plt.ylabel('Nombre total de médailles')
plt.title('Évolution du total de médailles par pays au fil des années (été)')
plt.legend(title='Pays', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

```python
medals_by_country_and_year = df_winter.groupby(['country_name', 'game_year']).size().unstack(fill_value=0)

top_countries = df['country_name'].value_counts().head(10).index

medals_by_country_and_year = medals_by_country_and_year.loc[top_countries]

plt.figure(figsize=(12, 8))
for country in medals_by_country_and_year.index:
    plt.plot(medals_by_country_and_year.columns, medals_by_country_and_year.loc[country], label=country)

plt.xlabel('Année')
plt.ylabel('Nombre total de médailles')
plt.title('Évolution du total de médailles par pays au fil des années (hiver)')
plt.legend(title='Pays', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

```python
splited_team_solo_summer = df_summer.groupby(['country_name', 'participant_type']).size().unstack(fill_value=0)

splited_team_solo_summer['Total'] = splited_team_solo_summer.sum(axis=1)
splited_team_solo_summer = splited_team_solo_summer.sort_values(by='Total', ascending=False).head(10)
splited_team_solo_summer.rename(columns={'Athlete': 'Individuel', 'GameTeam': 'Collectif'}, inplace=True)
ax = splited_team_solo_summer[['Individuel', 'Collectif']].plot(kind='bar', stacked=False, figsize=(12, 8))

plt.xlabel('Pays')
plt.ylabel('Nombre de médailles')
plt.title('Nombre de médailles des 10 pays ayant le plus de médailles par type d\'épreuve (été)')
plt.legend(title='Type de médaille', loc='upper right')
plt.grid(True)
plt.show()
```

```python
splited_team_solo_winter = df_winter.groupby(['country_name', 'participant_type']).size().unstack(fill_value=0)

splited_team_solo_winter['Total'] = splited_team_solo_winter.sum(axis=1)
splited_team_solo_winter = splited_team_solo_winter.sort_values(by='Total', ascending=False).head(10)
splited_team_solo_winter.rename(columns={'Athlete': 'Individuel', 'GameTeam': 'Collectif'}, inplace=True)
ax = splited_team_solo_winter[['Individuel', 'Collectif']].plot(kind='bar', stacked=False, figsize=(12, 8))

plt.xlabel('Pays')
plt.ylabel('Nombre de médailles')
plt.title('Nombre de médailles des 10 pays ayant le plus de médailles par type d\'épreuve (hiver)')
plt.legend(title='Type de médaille', loc='upper right')
plt.grid(True)
plt.show()
```

```python
count = df_summer['participant_type'].value_counts()
count = count.head(20)

plt.figure(figsize=(10, 6))
plt.bar(count.index, count.values, color='skyblue')
plt.xlabel('type d\'équipe')
plt.ylabel('Nombre de médailles possibles')
plt.title('Distribution des épreuve par type d\'équipe')
plt.xticks(rotation=90)
plt.show()
```

```python
total_medals_solo = df_summer['participant_type'].value_counts().values[0]
total_medals_team = df_summer['participant_type'].value_counts().values[1]

splited_team_solo_summer['solo_percentage'] = (splited_team_solo_summer['Individuel'] / total_medals_solo) * 100
splited_team_solo_summer['team_percentage'] = (splited_team_solo_summer['Collectif'] / total_medals_solo) * 100

ax = splited_team_solo_summer[['solo_percentage', 'team_percentage']].plot(kind='bar', stacked=False, figsize=(12, 8))

plt.xlabel('Pays')
plt.ylabel('% de médailles obtenu')
plt.title('Part des médailles obtenue des 10 pays ayant le plus de médailles par type d\'épreuve (été)')
plt.legend(title='Type de médaille', loc='upper right')
plt.grid(True)
plt.show()
```

```python
total_medals_solo = df_winter['participant_type'].value_counts().values[0]
total_medals_team = df_winter['participant_type'].value_counts().values[1]

splited_team_solo_winter['solo_percentage'] = (splited_team_solo_winter['Individuel'] / total_medals_solo) * 100
splited_team_solo_winter['team_percentage'] = (splited_team_solo_winter['Collectif'] / total_medals_solo) * 100

ax = splited_team_solo_winter[['solo_percentage', 'team_percentage']].plot(kind='bar', stacked=False, figsize=(12, 8))

plt.xlabel('Pays')
plt.ylabel('% de médailles obtenu')
plt.title('Part des médailles obtenue des 10 pays ayant le plus de médailles par type d\'épreuve (hiver)')
plt.legend(title='Type de médaille', loc='upper right')
plt.grid(True)
plt.show()
```
